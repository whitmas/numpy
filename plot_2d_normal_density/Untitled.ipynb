```python
import numpy as np
x = np.linspace(-1, 1, 3)
y = np.linspace(-2, 2, 3)
X, Y = np.meshgrid(x, y)
print(x)
print(y)
print(X)
print(Y)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(X.reshape((-1,)), Y.reshape((-1,)), 'o', markersize=15)
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-4, 4, 100)
y = np.linspace(-4, 4, 100)
X, Y = np.meshgrid(x, y)
Z = np.exp(-0.5*(X**2+Y**2)) / (2*np.pi)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D  # <-- Note the capitalization! 
fig = plt.figure(figsize=(9, 6))
ax = Axes3D(fig)  # <-- Note the difference from your original code...
ax.plot_surface(X, Y, Z,
                rstride=2,
                cstride=2,
                cmap=mpl.cm.coolwarm,
                linewidth=0.5,
                antialiased=True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```
