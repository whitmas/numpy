```python
x = np.linspace(-4, 4, 100)
y = np.linspace(-4, 4, 100)
X, Y = np.meshgrid(x, y)
Z = np.exp(-0.5*(X**2+Y**2)) / (2*np.pi)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D  # <-- Note the capitalization! 
fig = plt.figure(figsize=(9, 6))
ax = Axes3D(fig)  # <-- Note the difference from your original code...
ax.plot_surface(X, Y, Z,
                rstride=2,
                cstride=2,
                cmap=mpl.cm.coolwarm,
                linewidth=0.5,
                antialiased=True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```

Draw the following pdf:
$$
f({\bf x})=\frac{1}{2\pi|\Sigma|^{1/2}} exp\left(-\frac{1}{2}{\bf x}^T\Sigma^{-1}{\bf x}\right)
$$
where
$$
\Sigma=\left[\begin{array}{cc}
1&0.5\\
0.5&1
\end{array}\right]$$
and
$$
|\Sigma|=det(\Sigma)=1*1-0.5*0.5
$$

Draw the graph of following function:
$$
f(x,y)=x^2+y^2+\mbox{noize}
$$
where
$$
\mbox{noize}=0.3*z,\quad z\quad\mbox{iid standard normal}
$$

```python
x = np.arange(-1, 1, 0.05)
y = np.arange(-1, 1, 0.05)
X, Y = np.meshgrid(x, y)   
Z = X**2 + Y**2 + 0.3*np.random.normal(0., 1., X.shape)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D  # <-- Note the capitalization! 
fig = plt.figure(figsize=(9, 6))
ax = Axes3D(fig)  # <-- Note the difference from your original code...
ax.plot_surface(X, Y, Z,
                rstride=2,
                cstride=2,
                cmap=mpl.cm.coolwarm,
                linewidth=0.5,
                antialiased=True)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_xlim([-1.5, 1.5])
ax.set_ylim([-1.5, 1.5])
ax.set_zlim([-1.0, 2.5])
plt.show()
```
