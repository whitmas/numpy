```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_sample_image
china = load_sample_image("china.jpg")
print(china.shape)
plt.imshow(china)
plt.axis('off')
plt.show()
```

```python
china_flat = china.reshape((-1,))
china_flat0 = china.reshape((1, -1))
china_flat1 = china.reshape((-1, 1))
print(china_flat.shape)
print(china_flat0.shape)
print(china_flat1.shape)
plt.imshow(china_flat.reshape((427,640,3)))
plt.axis('off')
plt.show()
```

```python
china_part = china[150:220, 130:200]
china_part_2 = china[150:220:2, 130:200:2]
china_part_3 = china[150:220:3, 130:200:3]
print(china_part.shape)
print(china_part_2.shape)
print(china_part_3.shape)
fig, axes = plt.subplots(1, 3, figsize=(10,3))
axes[0].imshow(china_part)
axes[0].axis('off')
axes[1].imshow(china_part_2)
axes[1].axis('off')
axes[2].imshow(china_part_3)
axes[2].axis('off')
plt.show()
```

```python
china_part_channel_0 = china[150:220, 130:200, 0]
china_part_channel_1 = china[150:220, 130:200, 1]
china_part_channel_2 = china[150:220, 130:200, 2]
print(china_part_channel_0.shape)
print(china_part_channel_1.shape)
print(china_part_channel_2.shape)
fig, axes = plt.subplots(1, 3, figsize=(10,3))
axes[0].imshow(china_part_channel_0)
axes[0].axis('off')
axes[1].imshow(china_part_channel_1)
axes[1].axis('off')
axes[2].imshow(china_part_channel_2)
axes[2].axis('off')
plt.show()
```

```python
china_part_gray = china_part.mean(axis=2)
china_part_2_gray = china_part_2.mean(axis=2)
china_part_3_gray = china_part_3.mean(axis=2)
print(china_part_gray.shape)
print(china_part_2_gray.shape)
print(china_part_3_gray.shape)
fig, axes = plt.subplots(1, 3, figsize=(10,3))
axes[0].imshow(china_part_gray)
axes[0].axis('off')
axes[1].imshow(china_part_2_gray)
axes[1].axis('off')
axes[2].imshow(china_part_3_gray)
axes[2].axis('off')
plt.show()
```
