# Numerical Integral

$$\displaystyle\int_0^1 x^3 dx$$

```python
import numpy as np
def f(x): return x**3
n = 5  
x = np.linspace(0, 1, num=n, endpoint=False)
dx = x[1] - x[0]
y = f(x)
numerical_integral_value = np.sum(y*dx) 
print('nuber of subintervals : ', n)
print('integral value        : ', numerical_integral_value)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# figure of y = f(x)
x = np.linspace(0,1)
y = f(x)
plt.plot(x, y, 'r')

# figure of small boxes
for i in range(n):
    x_start = i*dx
    x_end = x_start + dx
    plt.plot([x_start,x_start],[0,f(x_start)],'--b')
    plt.plot([x_end,x_end],[0,f(x_start)],'--b')
    plt.plot([x_start,x_end],[0,0],'--b')
    plt.plot([x_start,x_end],[f(x_start),f(x_start)],'--b')
    
# show figure    
plt.show()
```

# Speed of Converegence

```python
import numpy as np

def f(x): return x**3

n_mesh = []
integral_value = []

for n in range(10,1000,50):
    
    x = np.linspace(0, 1, num=n, endpoint=False)
    dx = x[1] - x[0]
    y = f(x)
    numerical_integral_value = np.sum(y*dx)
    
    n_mesh.append(n)
    integral_value.append(numerical_integral_value)
```

```python
%matplotlib inline
import numpy as np
plt.plot(n_mesh, integral_value, '-*') 
plt.plot(n_mesh, 0.25*np.ones_like(n_mesh), '--r') 
plt.xlabel('Number of Subintervals')
plt.ylabel('Numerical Integration Value')
plt.ylim([0,0.26])
plt.show()
```
